### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
unique_values_per_column = my_heartdata.nunique()

# Finding the column with the maximum number of unique values
max_unique_column = unique_values_per_column.idxmax()
max_unique_value_count = unique_values_per_column.max()

# Printing the number of unique values for each column
print("Number of unique values per column:")
print(unique_values_per_column)

# Printing the column with the maximum number of unique values
print(f"\nThe column with the most diversity is '{max_unique_column}' with {max_unique_value_count} unique values.")
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import matplotlib.pyplot as plt

# Assuming Mental_health is your DataFrame
numeric_column = 'chol'

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(my_heartdata[numeric_column].dropna(), bins=30, color='orange', edgecolor='black')
plt.title(f'Distribution of {numeric_column}')
plt.xlabel(numeric_column)
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.scatterplot(x='thalach', y='age', data=my_heartdata, color='darkblue', alpha=0.9, edgecolor='black')
plt.xticks(rotation=45, ha='right')
plt.title('Scatter Plot: Relationship Between Age And Max Heart Rate Achieved', fontsize=14)
plt.xlabel('Max Heart Rate Achieved (thalach)', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
